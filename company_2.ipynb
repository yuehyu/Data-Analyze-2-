```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df1=pd.read_csv('FABExptest.csv')
df1.drop(['date','月份','星期','淡旺季','AI啟動','LeadTime','QAIndex','深圳總產能下之總費用'],axis='columns',inplace=True)
df1
```

```python
df1=df1.rename(columns={'龍華一廠':'LongWhua1','龍華二廠':'LongWhua2','龍華三廠':'LongWhua3','LCM廠':'LCM','觀瀾廠':'GuanFactory',
                            '模具部':'Module','製工部':'ME','製程部':'MP','AI部':'AIDep',
                            'QA部':'QADep','總經理室':'GM',
                            '財務部':'FinanceDep','管理部':'ManageDep','業務一部':'Sales1',
                            '業務二部':'Sales2'})
```

```python
display(df1.info())
display(df1.corr())  
# data.corr() 表示data兩個數據之間的相關性
```

```python
fig,ax=plt.subplots(figsize=(20,9))
sns.heatmap(pd.DataFrame(df1.corr()),annot=True,vmax=0,xticklabels=True,yticklabels=True,square=False,cmap="YlGnBu")
ax.set_title("Heatmap of Corr",fontsize=18)
ax.set_ylabel("Site Names",fontsize=12)
ax.set_xlabel("Site Names",fontsize=12)
plt.savefig('output.png')
plt.show()
```
